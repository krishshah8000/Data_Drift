```python
import numpy as np
from scipy.spatial.distance import jensenshannon

# Training data
expected = np.array([22,25,28,30,33,35,37,40,42,45])

# New data
actual = np.array([25,30,35,38,40,45,48,50,55,60])

# Create common bins
bins = [20,30,40,50,60]

# Count values
expected_count, _ = np.histogram(expected, bins)
actual_count, _ = np.histogram(actual, bins)

# Convert to probabilities
P = expected_count / expected_count.sum()
Q = actual_count / actual_count.sum()

# Jensen-Shannon Distance
js_distance = jensenshannon(P, Q, base=2)

print("Training Distribution :", P)
print("New Distribution      :", Q)
print("Jensen-Shannon Distance =", js_distance)

# Interpretation
if js_distance < 0.10:
    print("Very similar distributions")
elif js_distance < 0.20:
    print("Small distribution shift")
elif js_distance < 0.40:
    print("Moderate distribution shift")
else:
    print("Large distribution shift")
```
