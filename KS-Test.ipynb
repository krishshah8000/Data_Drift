```python
import numpy as np
from scipy.stats import ks_2samp

# Training Data
expected = np.array([
    22,25,28,30,33,
    35,37,40,42,45
])

# New Data
actual = np.array([
    25,30,35,38,40,
    45,48,50,55,60
])

# Perform KS Test
ks_statistic, p_value = ks_2samp(expected, actual)

print("KS Statistic :", ks_statistic)
print("P-value      :", p_value)

# Interpretation
if p_value > 0.05:
    print("The two datasets likely come from the same distribution.")
else:
    print("The two datasets likely come from different distributions.")
```
