```python
import numpy as np
import pandas as pd


expected = np.array([
    22,25,28,30,33,35,37,
    40,42,45,48,50,55,58,
    60,65,70,75,80,85
])



actual = np.array([
    25,30,35,38,
    40,45,48,50,55,
    60,65,70,72,75,
    80,82,85,88,90,92
])


bins = [20,40,60,80,100]


expected_count, _ = np.histogram(expected, bins)
actual_count, _ = np.histogram(actual, bins)



expected_pct = expected_count / len(expected)
actual_pct = actual_count / len(actual)

# Avoid divide-by-zero
expected_pct = np.where(expected_pct == 0, 0.0001, expected_pct)
actual_pct = np.where(actual_pct == 0, 0.0001, actual_pct)


psi = (actual_pct - expected_pct) * np.log(actual_pct / expected_pct)


df = pd.DataFrame({
    "Bin": ["20-40","40-60","60-80","80-100"],
    "Expected Count": expected_count,
    "Actual Count": actual_count,
    "Expected %": expected_pct,
    "Actual %": actual_pct,
    "PSI": psi
})

print(df)

print("\nTotal PSI =", psi.sum())


total_psi = psi.sum()

if total_psi < 0.1:
    print("Stable population")
elif total_psi < 0.25:
    print("Moderate drift")
else:
    print("Significant drift")
```
